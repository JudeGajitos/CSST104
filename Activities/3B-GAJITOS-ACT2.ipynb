# **1. Import Libraries**

```python
import pandas as pd # Data Manipulation
import numpy as np # Mathematical Operations
import matplotlib.pyplot as plt # Data Visualization
import seaborn as sne
from sklearn.linear_model import LinearRegression # Training Data/Data Modeling
```

# **2. Loading Dataset**

```python
lesson4 = pd.read_csv('Lesson_4_student_grades_dataset.csv')
print(lesson4.head())
```

# **3. Visualizing the Data**

```python
plt.figure(figsize = (14,7))
plt.scatter(lesson4['Hours_of_Study'], lesson4['Grades'])
plt.xlabel('Hours_of_Study (x)')
plt.ylabel('Grades (y)')
plt.title("Scatter Plot of Hours of Study vs Grades")

plt.figure(figsize = (14, 7))
plt.scatter(lesson4['Previous_Exam_Scores'], lesson4['Grades'])
plt.xlabel('Previous_Exam_Scores (x)')
plt.ylabel('Grades (y)')
plt.title('Scatter Plot of Previous Exam Score vs Grades')
plt.show()
```

# **4. Preparing Data for Modeling**

```python
x = lesson4[['Previous_Exam_Scores']]
y = lesson4[['Grades']]
```

# **5. Fitting the Linear Regression Model**

```python
model = LinearRegression()
model.fit(x,y)
```

# **6. Getting Model Coefficients**

```python
slope = model.coef_[0]
intercept = model.intercept_
print("Slope:", slope)
print("Intercept:", intercept)
```

# **7. Visualizing the Linear Regression Line:**

```python
plt.figure(figsize = (16, 9))
plt.scatter(x,y)
plt.xlabel('Previous_Exam_Scores (x)')
plt.ylabel('Grades (y)')
plt.plot(x, model.predict(x), color = 'orange')
plt.show()
```
