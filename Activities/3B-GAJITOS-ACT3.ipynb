```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
lesson4 = pd.read_csv('Lesson_4_student_grades_dataset.csv')
lesson4.head()
```

```python
x = pd.DataFrame(lesson4, columns = ['Hours_of_Study', 'Attendance_Percentage', 'Previous_Exam_Scores'])
y = lesson4['Grades']

X_train, X_test, y_train, y_test, = train_test_split(x,y, test_size = 0.2, random_state = 42)
x.head(10)
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_predict = model.predict(X_test)
print(X_test)
print(y_predict)
```

```python
mse = mean_squared_error(y_test, y_predict)
print("Mean Squared Error:", mse)
```

```python
new_grades = [[2.4, 94.54, 84.5]]
predicted_grades = model.predict(new_grades)
print(f"Predicted: {predicted_grades[0]}")
```
