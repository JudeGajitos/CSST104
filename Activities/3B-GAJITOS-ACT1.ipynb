# **1. Loading Dataset**

```python
import pandas as pd

df = pd.read_csv('hardwareStore.csv')
print(df.isnull().sum())
```

# **2. Initial Data Analysis**

```python
import matplotlib.pyplot as plt
import seaborn as sne
summary_stat = df.describe()
product_category = df.groupby('CATEGORY_NAME')['CATEGORY_NAME'].count()
print(summary_stat)
print(product_category)
```

# **3. Category Analysis**

```python
product_category.plot(kind = 'bar', figsize = (12, 6))
plt.xlabel('CATEGORY_NAME')
plt.title('Distribution of Products across Categories')
plt.xticks(rotation = 90)
plt.tight_layout()
plt.show()
highest_category_product = product_category.idxmax()

print("Highest number of Product is:", highest_category_product)
```

# **4. Cost and Price Analysis**

```python
ave_standard_cost = df['STANDARD_COST'].mean
ave_list_price = df['LIST_PRICE'].mean
print(ave_standard_cost)
print(ave_list_price)

highest_standard_cost = df['STANDARD_COST'].idxmax()
highest_list_price = df['LIST_PRICE'].idxmax()
print("Highest Standard Cost:",highest_standard_cost)
print("Highest List Price:",highest_list_price)
```

# **5. Location Analysis**

```python
quantity_in_location = df.groupby('CITY')['QUANTITY'].idxmax()
quantity_in_location.plot(kind = 'bar', figsize = (12, 6))
plt.xlabel('CITY NAME')
plt.title('Distribution of Products in Cities')
plt.xticks(rotation = 45)
plt.tight_layout()
plt.show()
print(quantity_in_location)
```

# **6. Insight and Reporting**

```python
report = """
analysis Report
1. Data Loading and Initial Analysis:
  Basic Statistics of the Dataset:
  {}
2. Category Analysis:
  Unique Categories and their Counts:
  {}
  Highest Number of Product:{}
3. Cost and Price Analysis:
  Average Standard Cost: {}
  Average List Price: {}
  Highest Standard Cost: {}
  Highest List Price: {}
4. Location Analysis:
  Highest Quantity of Product in Stock by Location: {}
""". format(summary_stat, product_category, highest_category_product, ave_standard_cost, ave_list_price, highest_standard_cost, highest_list_price, quantity_in_location)
print(report)
```
