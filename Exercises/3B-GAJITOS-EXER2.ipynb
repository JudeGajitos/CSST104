# **1. Creating Dataset**

```python
import pandas as pd

data = {
    'student_id': range(1, 16),
    'gender': ["Male", "Male", "Female", "Male", "Female", "Female", "Female", "Male", "Female", "Male", "Male", "Male", "Male", "Female", "Female"],
    'test_date': pd.date_range(start = "2024-02-14", periods = 15),
    'study_hours': [2, 5, 4, 2, 2, 3, 1, 1, 4, 5, 5, 3, 2, 6, 3],
    'test_score': [82, 89, 86, 80, 83, 84, 78, 79, 88, 92, 93, 84, 81, 95, 84],
    'attendance': ["Absent", "Present", "Present", "Present", "Absent", "Absent", "Absent", "Present", "Absent", "Present", "Absent", "Present", "Absent", "Present", "Absent"]
}

df = pd.DataFrame(data)
df.to_csv('student_performance.csv', index= False)
```

# **2. Data Loading and Initial Examination**

```python
df = pd.read_csv('student_performance.csv')
print(df.head(7))
print(df.isnull().sum())
```

# **3. Exploratory Data Analysis**

```python
import matplotlib.pyplot as plt
import seaborn as sne

summary_stats = df[['test_score', 'study_hours']].describe()

#Histogram
plt.figure(figsize = (12,6))
plt.subplot(1,2,1)
sne.histplot(df['test_score'], bins = 15, kde = True)
plt.title("Distribution of Test Score")

#Histogram
plt.subplot(1,2,2)
sne.histplot(df['study_hours'], bins = 15, kde = True)
plt.title("Distribution of Study Hours")

#ScatterPlot
plt.figure(figsize = (10, 5))
sne.scatterplot(data = df, x = 'study_hours', y = 'test_score')
plt.title("Scatter Plot of Study Hours vs. Test Score")
```

# **4. Data Visualization**

```python
plt.figure(figsize = (5, 4))
sne.countplot(data = df, x = 'gender')
plt.title("Gender Distribution")

plt.figure(figsize = (5, 4))
sne.countplot(data = df, x = 'attendance')
plt.title("Attendance Distribution")

plt.figure(figsize = (5, 4))
sne.lineplot(data = df, x = 'test_date', y = 'test_score')
plt.title("Trend in Test Score Over Time")
plt.xticks(rotation = 75)

correlation_matrix = df[['study_hours', 'test_score', 'attendance']].corr()
plt.figure(figsize = (5,4))
sne.heatmap(correlation_matrix, annot = True, cmap = 'coolwarm')
plt.title("Correlation Analysis")
```
