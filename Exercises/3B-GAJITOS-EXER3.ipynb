# **1. Import Libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

# **2. Creating a Synthetic Dataset**

```python
np.random.seed()
x = np.random.rand(100, 3)
y = x[:,0] * 15000 + x[:,1] * 100000 + x[:,2] * 50000 + 50000
```

# **3. Converting to a DataFrame**

```python
data = pd.DataFrame(x, columns = ['Size', 'Bedrooms', 'Age'])
data['Price'] = y

X_train, X_test, y_train, y_test, = train_test_split(x,y, test_size = 0.2, random_state = 42)
data.head()
```

# **4. Initializing the Model**

```python
model = LinearRegression()
```

# **5. Fitting the Model**

```python
model.fit(X_train, y_train)
```

# **6. Making Prediction**

```python
y_predict = model.predict(X_test)
print(X_test)
print(y_predict)
```

# **7. Calculating the Mean Squared Error**

```python
mse = mean_squared_error(y_test, y_predict)
print("Mean Squared Error:", mse)
```

# **8. Predicting the Price of a New House**

```python
new_house = [[0.5, 2, 10]]
predicted_price = model.predict(new_house)
print(f"Predicted: {predicted_price[0]}")
```
